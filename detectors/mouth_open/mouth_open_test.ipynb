# Mouth Open Test

```python
import glob
import pandas as pd
from datetime import datetime
from mouth_open_tests import is_mouth_closed

all_images = glob.glob("D:/oneid_backend/icao-guidelines-server/dataset/*.JPG")
```

```python
def logger(image, is_valid, path="./"):
    time_str = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    df_headers = pd.DataFrame({
        'log_time': [time_str],
        'image': [image],
        'is_valid': [is_valid],
    })
    df_headers.to_csv(f'{path}mouth_close_logs.csv', mode='a', index=False, header=False)
```

```python
# time duration
start_time = datetime.now()

for image_path in all_images:
    check = is_mouth_closed(image_path)
    logger(image_path, check)
    print(image_path, check)
```
